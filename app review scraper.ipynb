```python
import pandas as pd
import numpy as np
import plotly_express as px
```

```python
from google_play_scraper import app, Sort, reviews_all
fishbowl = reviews_all('com.fishbowlmedia.fishbowl',sleep_milliseconds=0,lang='en',country='in',sort=Sort.NEWEST)
```

```python
#fishbowl
```

```python
df = pd.json_normalize(fishbowl)
```

```python
df.head(3)
```

```python
df['score'].mean()
#df['score'].value_counts
```

```python
df['reviewCreatedVersion'].value_counts()
```

```python
from transformers import pipeline
sentiment_analysis = pipeline("sentiment-analysis",model="siebert/sentiment-roberta-large-english")
print(sentiment_analysis("I love this!"))
```

```python
df.dtypes
```

```python
df['content'] = df['content'].astype('str')
```

```python
df.shape
```

```python
df500 = df.head(500)
df500.shape
```

```python
df500['result'] = df['content'].apply(lambda x:sentiment_analysis(x))
```

```python
df500
```

```python
df500.shape
```

```python
df500['sentiment'] = df500['result'].apply(lambda x: (x[0]['label']))
df500['score'] = df500['result'].apply(lambda x: (x[0]['score']))
```

```python
df500.head()
```

```python
df500['score'].mean()
```

```python
df500['sentiment'].value_counts(normalize=True)
```

```python
fig = px.histogram(df500, x = 'sentiment', color = 'sentiment', text_auto= True)
fig.show()
```

```python
df500.to_csv('fishbowl_sentiment.csv')
```
